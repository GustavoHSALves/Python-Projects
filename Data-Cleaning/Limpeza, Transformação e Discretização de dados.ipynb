```python
# Preenchendo com a média
#df.fillna(df.mean())
# Preenchendo com a média
# First convert numeric columns to numeric type, then fill NaN values
# For numeric columns only
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
df
# For string columns (if needed)
# string_cols = df.select_dtypes(include=['object']).columns
# df[string_cols] = df[string_cols].fillna(df[string_cols].mode().iloc[0])
```

```python
# Preenchendo com a moda
df.fillna(df.mode().iloc[0])
```

```python
# Preenchendo com a mediana
#df.fillna(df.median().iloc[0])

numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
df
```

```python
df['year'] = df['Reported Date'].map(lambda x: str(x)[0:4] if x is not np.nan else np.nan)

df['year']
```

```python
df['day'] = df['Reported Date'].map(lambda x: int(str(x)[-2:]) if x is not np.nan else np.nan)
df['day'] 
```

```python
bins = [0, 10, 20, 31]
periodo_mes = pd.cut(df['day'].to_numpy(), bins=bins)
```

```python
df['periodo_mes'] = periodo_mes
```
